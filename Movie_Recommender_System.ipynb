```python
import pandas as pd 
import numpy as np 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
newmovies = movies.merge(credits,on='title')
```

```python
newmovies.info()
```

```python
mainpage = newmovies[['movie_id','title','overview','genres', 'keywords','cast','crew']]
mainpage
```

```python
mainpage.head(2)
```

```python
# genres
# id (for html)
# keywords
# title
# overview
#  cast
# crew
```

```python
mainpage.isnull().sum()
```

```python
mainpage.dropna(inplace=True)
```

```python
mainpage.duplicated().sum()
```

```python
mainpage.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
mainpage['genres'] = mainpage['genres'].apply(convert)
```

```python
mainpage.head(1)
```

```python
mainpage['keywords'] = mainpage['keywords'].apply(convert)
```

```python
mainpage.head(1)
```

```python
import ast
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
mainpage['cast'] = mainpage['cast'].apply(convert3)
```

```python
import ast
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
mainpage['crew'] = mainpage['crew'].apply(fetch_director)
```

```python
mainpage
```

```python
mainpage['overview'] = mainpage['overview'].apply(lambda x:x.split())
```

```python
mainpage
```

```python
mainpage['genres'] = mainpage['genres'].apply(lambda x:[i.replace(" ","") for i in x ])
mainpage['keywords'] = mainpage['keywords'].apply(lambda x:[i.replace(" ","") for i in x ])
mainpage['cast'] = mainpage['cast'].apply(lambda x:[i.replace(" ","") for i in x ])
mainpage['crew'] = mainpage['crew'].apply(lambda x:[i.replace(" ","") for i in x ])
```

```python
mainpage
```

```python
mainpage['tags'] = mainpage['overview'] + mainpage['genres'] + mainpage['keywords'] + mainpage['cast'] + mainpage['crew']
```

```python
mainpage
```

```python
new_df = mainpage[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(1)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simlarty = cosine_similarity(vectors)
```

```python
sorted(simlarty[0], reverse=True)
```

```python
sorted(list(enumerate(simlarty[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
simlarty
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = simlarty[movie_index] 
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
